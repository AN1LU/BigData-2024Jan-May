```r
# Función para realizar análisis descriptivo y gráfico de un dataframe en R
analyze_dataframe <- function(df) {

  # Categorizar las variables en cualitativas y cuantitativas
  categorical_vars <- sapply(df, is.factor)
  numeric_vars <- sapply(df, is.numeric)

  # Análisis descriptivo para variables categóricas
  summary_categorical <- lapply(df[categorical_vars], table)

  # Análisis descriptivo para variables numéricas
  summary_numeric <- lapply(df[numeric_vars], summary)

  # Gráficos para variables categóricas
  for (var in names(summary_categorical)) {
    barplot(summary_categorical[[var]], main = paste("Barplot of", var))
  }

  # Gráficos para variables numéricas
  for (var in names(summary_numeric)) {
    hist(df[[var]], main = paste("Histogram of", var))
  }

  # Gráfico de dispersión por pares
  pairs(df[numeric_vars])
}

# Ejemplo de uso
data(iris)  # Usando el dataset iris como ejemplo
analyze_dataframe(iris)
```
