Como bien sabemos un data frame es definido como tablas donde cada columna puede tener un tipo diferente pero cada columna debe contener el mismo número de datos (longitud).

Estas estructuras de programación son muy utilizadas en big data para manejar distintas colecciones de datos, ya sea que provengan de un archivo CSV, TXT, JSON, etc.


De acuerdo con la Clase R Studio se dice que para utilizar un Data Frame en R es necesario seguir la siguiente estructura:


`nombre_del_dataframe <- data.frame(columna_tipo_string=c("Contenido tipo cadena del dataframe"), columna_2(34,35,6,7))`
